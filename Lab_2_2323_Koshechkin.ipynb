В качестве датасета были использованы данные с https://www.kaggle.com/datasets/jamiedcollins/hjsjdjdjdjd?resource=download 

Этот набор данных позволяет глубоко изучить финансовые показатели Apple за один из самых знаковых периодов в истории компании - с 2009 по 2014 год. В этот период компания выросла из технологического лидера в глобального техногиганта.

Этот набор данных содержит следующие атрибуты:
- EBITDA (millions) – Прибыль до вычета процентов, налогов, износа и амортизации.
- Revenue (millions) – Общая выручка компании.
- Gross Profit (millions) – Валовая прибыль (выручка минус прямые затраты).
- Op Income (millions) – Операционная прибыль (прибыль от основной деятельности).
- Net Income (millions) – Чистая прибыль (после всех расходов и налогов).
- EPS – Прибыль на акцию.
- Shares Outstanding – Количество акций в обращении.
- Year Close Price – Цена закрытия акций на конец года.
- Total Assets (millions) – Общая стоимость активов компании.
- Cash on Hand (millions) – Денежные средства компании.
- Long Term Debt (millions) – Долгосрочные обязательства компании.
- Total Liabilities (millions) – Общие обязательства компании.
- Gross Margin – Валовая маржа (отношение валовой прибыли к выручке).
- PE ratio – Коэффициент цена/прибыль (отношение цены акции к прибыли на акцию).
- Employees – Количество сотрудников компании.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


#Загрузка таблицы с очищенными данными (убраны $,в числах запятые заменены на точки)
data = pd.read_csv("Cleaned_data_Apple.csv")
data
```

```python
#Добавим столбец с аттрибтом, в которым посчитаем количество прибыли компании на 1 сотрудника:
data['Revenue per employee'] = round(data['Revenue (millions)'] / data['Employees'] * 1000000, 2)
data
```

```python
#Проверим на выбросы и отстутсвие значений:
# Анализ каждого атрибута
columns = data.columns.drop("year")  # Исключаем год из анализа

for col in columns:
    print(f"\n\nАтрибут: {col}")
    print(f"Среднее значение: {data[col].mean():.2f}")
    print(f"Стандартное отклонение: {data[col].std():.2f}")
    
    # Проверка пропущенных значений
    missing_count = data[col].isna().sum()
    print(f"Пропущенные значения: {missing_count}")
     
    # Выбросы: межквартильный размах (IQR)
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
        # Нахождение выбросов для текущего столбца
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
        
     # Обработка выбросов
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    if not outliers.empty:
        print(f"Выбросы для {col}:\n{outliers[['year', col]]}")
        mean_value = data[col].mean()
        # Замена выбросов на среднее значение
        data[col] = data[col].apply(
            lambda x: mean_value if x < lower_bound or x > upper_bound else x
        )
        print(f"Выбросы заменены средним значением: {mean_value:.2f}")
    else:
        print(f"Выбросы не найдены для {col}.")
    
```

```python
data
```

```python
# Импортируем необходимые библиотеки
from sklearn.cluster import KMeans
import seaborn as sns

# Задание 3 Построение графиков зависимости и описание кластеров
plt.figure(figsize=(10, 6))
plt.scatter(data["Revenue (millions)"], data["EBITDA (millions)"], c='blue', label="Revenue vs EBITDA")
plt.xlabel("Revenue (millions)")
plt.ylabel("EBITDA (millions)")
plt.title("Зависимость EBITDA от Revenue")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(data["Revenue (millions)"], data["Net Income (millions)"], c='green', label="Revenue vs Net Income")
plt.xlabel("Revenue (millions)")
plt.ylabel("Net Income (millions)")
plt.title("Зависимость Net Income от Revenue")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(data["EBITDA (millions)"], data["Net Income (millions)"], c='purple', label="EBITDA vs Net Income")
plt.xlabel("EBITDA (millions)")
plt.ylabel("Net Income (millions)")
plt.title("Зависимость Net Income от EBITDA")
plt.legend()
plt.grid()
plt.show()

# Зависимость Total Assets от Long Term Debt
plt.figure(figsize=(10, 6))
plt.scatter(data["Total Assets (millions)"], data["Long Term Debt (millions)"], c='red', label="Total Assets vs Long Term Debt")
plt.xlabel("Total Assets (millions)")
plt.ylabel("Long Term Debt (millions)")
plt.title("Зависимость Long Term Debt от Total Assets")
plt.legend()
plt.grid()
plt.show()
```

```python
# Задание 4 Применение метода KMeans
# Выбор параметров для кластеризации
X = data[["Revenue (millions)", "Net Income (millions)"]]

# Применение KMeans
kmeans = KMeans(n_clusters=3, random_state=0)
data["Cluster"] = kmeans.fit_predict(X)

# Визуализация кластеров
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=data["Revenue (millions)"],
    y=data["Net Income (millions)"],
    hue=data["Cluster"],
    palette="viridis",
    s=100
)
plt.title("Кластеры по Revenue и Net Income")
plt.xlabel("Revenue (millions)")
plt.ylabel("Net Income (millions)")
plt.grid()
plt.show()

# Вывод центров кластеров
print("Координаты центров кластеров:")
print(kmeans.cluster_centers_)

# KMeans: Кластеры по EBITDA и Revenue
X_ebitda_revenue = data[["EBITDA (millions)", "Revenue (millions)"]]

# Применение KMeans
kmeans_ebitda_revenue = KMeans(n_clusters=3, random_state=42)
data["Cluster_EBITDA_Revenue"] = kmeans_ebitda_revenue.fit_predict(X_ebitda_revenue)

# Визуализация кластеров
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=data["EBITDA (millions)"],
    y=data["Revenue (millions)"],
    hue=data["Cluster_EBITDA_Revenue"],
    palette="coolwarm",
    s=100
)
plt.title("Кластеры по EBITDA и Revenue")
plt.xlabel("EBITDA (millions)")
plt.ylabel("Revenue (millions)")
plt.grid()
plt.show()

# KMeans: Кластеры по Total Assets и Long Term Debt
X_assets_debt = data[["Total Assets (millions)", "Long Term Debt (millions)"]]

# Применение KMeans
kmeans_assets_debt = KMeans(n_clusters=3, random_state=42)
data["Cluster_Assets_Debt"] = kmeans_assets_debt.fit_predict(X_assets_debt)

# Визуализация кластеров
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=data["Total Assets (millions)"],
    y=data["Long Term Debt (millions)"],
    hue=data["Cluster_Assets_Debt"],
    palette="viridis",
    s=100
)
plt.title("Кластеры по Total Assets и Long Term Debt")
plt.xlabel("Total Assets (millions)")
plt.ylabel("Long Term Debt (millions)")
plt.grid()
plt.show()

# Координаты центров кластеров
print("Координаты центров кластеров для EBITDA и Revenue:")
print(kmeans_ebitda_revenue.cluster_centers_)

print("\nКоординаты центров кластеров для Total Assets и Long Term Debt:")
print(kmeans_assets_debt.cluster_centers_)

```
